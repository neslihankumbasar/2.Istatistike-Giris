```python
### Soru 1 :
## dağılım = np.random.dağılım_adı([],[]) kodunu kullanarak, dağılım_adı yazan yere 
## seçtiğiniz bir dağılımı ve parantez içinde [] şeklinde gösterilen boşlukları yine kendi seçiminiz doğrultusunda doldurun. 
## Daha sonra, 
## a) Verinin histogramını çiziniz 
## b) Q-Q grafiğini çizip yorumlayınız
```

```python
import numpy as np
import matplotlib.pyplot as plt
yas = [18, 25, 22, 34, 26, 27, 19, 28, 30, 24, 45, 53, 42, 23, 57]
normal_dagilimli_yas=np.random.normal(np.mean(yas),np.std(yas),15)
yas.sort()
normal_dagilimli_yas.sort()
```

```python
## a) Verinin histogramı 
plt.figure(figsize=(8,4), dpi=100)
plt.hist(normal_dagilimli_yas, bins=20)
plt.show()
```

```python
## b) Q-Q grafiğini
plt.figure(figsize=(8,5), dpi = 200)
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font = {'family':'arial' ,'color':'darkblue', 'weight':'bold', 'size':10}
plt.title("Ornek Yas & Normal Dagilimli Yas Veri Kumesi", fontdict = baslik_font)
plt.xlabel("Ornek Yas Veri Kumesi", fontdict=eksen_font)
plt.ylabel("Normal Dagilimli Yas Veri Kumesi", fontdict=eksen_font)
plt.scatter(normal_dagilimli_yas, yas)
plt.plot([np.min(normal_dagilimli_yas), np.max(normal_dagilimli_yas)],
         [np.min(yas), np.max(yas)], c='red')
plt.show()

## Saga carpik bir grafik cikiyor
```

```python
### Soru 2 :
## https://finance.yahoo.com/ sitesini kullanarak google şirketinin 2015-2017 yılları arasında hisse senedi fiyatlarını elde edin. 
## Bu veriyi kullanarak hisse senedi getirisini şu formülle hesaplayın:

## (burada  p(t)  t zamanındaki hisse senedi fiyatını ve  p(t-1) ise t-1 zamanındaki hisse senedi fiyatını vermektedir). 

## Getiriler dağılımı hangi dağılıma daha yakındır?
```

```python
## Hem histogram hem de QQ ile dagilimini gorebiliriz:
## Normal bir dagilima sahip olmadigini goruyoruz.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
google_hisse = pd.read_csv('GOOG.csv')
google_hisse.sort_values(["Date"] , axis=0, ascending=True, inplace=True) 
google_liste = []
for i in range(1, len(google_hisse)):
    google_liste.append ( (google_hisse.loc[i, 'Close'] - google_hisse.loc[i-1, 'Close']) / google_hisse.loc[i-1, 'Close'])
plt.hist(google_liste)
plt.show()
```

```python
normal_dagilimli_veri = np.random.normal(np.mean(google_liste),
                                         np.std(google_liste), len(google_liste))
google_liste.sort()
normal_dagilimli_veri.sort()
plt.figure(figsize=(8,5), dpi = 200)
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
plt.title('Örnek Veri Kümesi & Normal Dağılımlı Veri Kümesi', fontdict = baslik_font)
plt.xlabel('Örnek Veri Kümesi', fontdict = eksen_font)
plt.ylabel('Normal Dağılımlı Veri Kümesi', fontdict = eksen_font)
plt.scatter(normal_dagilimli_veri, google_liste)
plt.plot([np.min(normal_dagilimli_veri), np.max(normal_dagilimli_veri)],
         [np.min(google_liste), np.max(google_liste)], c='red')
plt.show()
```
