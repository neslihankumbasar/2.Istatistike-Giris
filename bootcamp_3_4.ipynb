```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
%matplotlib notebook

# 1000 kere zar atalım
n = 1000

ortalama = []
for i in range(1,n):
    a = np.random.randint(1,7,10)
    ortalama.append(np.average(a))
    
# 10 örneklem seçimi
ortalama[1:10]

def clt(current):
    plt.cla()
    if current == 1000: 
        a.event_source.stop()

    plt.hist(ortalama[0:current])

    plt.gca().set_title('Zar Atışından Beklenen Değer')
    plt.gca().set_xlabel('Zar Atışının Ortalaması')
    plt.gca().set_ylabel('Dağılım')

    plt.annotate('Zar Atışı= {}'.format(current), [3,27])
    
fig = plt.figure()
ani = animation.FuncAnimation(fig, clt, interval=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8,5))
pop1=np.random.binomial(10,0.2,10000)
pop2=np.random.binomial(10,0.5,10000)

plt.hist(pop1,alpha=0.5, label='Populasyon 1')
plt.hist(pop2,alpha=0.5, label='Populasyon 2')

plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(8,5))

orneklem1=np.random.choice(pop1, 100, replace=True)
orneklem2=np.random.choice(pop2, 100, replace=True)

plt.hist(orneklem1, alpha=0.5 , label='Orneklem 1')
plt.hist(orneklem2, alpha=0.5 , label='Orneklem 2')

plt.legend(loc='upper right')
plt.show()
```

```python
print("1. Orneklemin ortalamasi : ", orneklem1.mean())
print("2. Orneklemin ortalamasi : ", orneklem2.mean())
print("1. Orneklemin standart sapmasi : ", orneklem1.std())
print("2. Orneklemin standart sapmasi : ", orneklem2.std())
```

```python
fark=orneklem2.mean()-orneklem1.mean()
print(fark)
```

```python
boyut=np.array([len(orneklem1), len(orneklem2)])
boyut=np.array([len(orneklem1), len(orneklem2)])
sd= np.array([orneklem1.std(), orneklem2.std()])
sd= np.array([orneklem1.std(), orneklem2.std()])

fark_se=  (sum(sd**2/boyut))**0.5
fark_se = (sum(sd ** 2 / boyut)) ** 0.5 
print(fark/fark_se)
```

```python
from scipy.stats import ttest_ind
ttest_degeri = ttest_ind(orneklem1, orneklem2, equal_var=False)
print(ttest_degeri)
```

```python
orneklem=[]
orneklem_ortalamalari=[]
for atis_sayisi in range(1,2000):
    atis=np.random.randint(low=1, high=7 ,size=1)[0]
    orneklem.append(atis)
    orneklem_ortalamalari.append(np.mean(orneklem))    
```

```python
def orneklem_plotla(orneklem_ortalaması,orneklem_sayisi, subplot_no):
    baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
    eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
    plt.subplot(subplot_no)
    plt.hlines(3.5, 0, orneklem_sayisi, colors= 'r' )
    plt.plot(orneklem_ortalaması[0:orneklem_sayisi])
    plt.ylim((1,6))
    plt.title("Örneklem Sayısı = {}".format(orneklem_sayisi), fontdict = baslik_font)
    plt.xlabel("Örneklem Sayısı", fontdict = eksen_font)
    plt.ylabel("Örneklem Ortalaması", fontdict = eksen_font)
    return plt
```

```python
plt.figure(figsize=(15,3), dpi=100)
orneklem_plotla(orneklem_ortalamalari,100, 131) 
orneklem_plotla(orneklem_ortalamalari, 500, 132)
orneklem_plotla(orneklem_ortalamalari, 2000, 133)
plt.show()
```
